```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
covid19_byTown = pd.read_csv('COVID-19_ByTown.csv') 
covid19_byTown
```

```python
hartford_city = covid19_byTown.loc[covid19_byTown['Town number'] == 64]
hartford_city
```

```python
hartford_city['Confirmed cases']
```

```python
HC_dailyCase = hartford_city[['Last update date','Confirmed cases']]
HC_dailyCase
```

```python
#change Nan to 0
HC_dailyCase = HC_dailyCase.fillna(0)
```

```python
#convert date type
HC_dailyCase['Last update date'] = pd.to_datetime(HC_dailyCase['Last update date'], format = '%m/%d/%Y')
HC_dailyCase
```

```python
HC_dailyCase['Last update date'].dtypes
```

```python
#Caculate Daily Cases
HC_dailyCase['Daily Confirmed Cases']= HC_dailyCase['Confirmed cases'].diff()
```

```python
#convert Nan to 0
HC_dailyCase = HC_dailyCase.fillna(0)
HC_dailyCase
```

```python
HC_dailyCase.to_csv("./HC_dailyCase.csv") 
```

```python
#set Date as index
HC_dailyCase_indexDate = HC_dailyCase.set_index('Last update date')
```

```python
HC_daily_rate = plt.figure(figsize =(15,6))
HC_daily_rate = HC_dailyCase_indexDate['Daily Confirmed Cases'].plot()
plt.xlabel("Date YYYY-MM")
plt.ylabel("Confirmed_Daily_Cases")
plt.legend()
plt.title('Confirmed Daily Case of COVID-19-Hartford Town,Connecticut')
```

```python
mask_HC = (HC_dailyCase['Last update date'] >= '2020-06-01') &(HC_dailyCase['Last update date'] <= '2021-05-31') 
HC_cases_mask=HC_dailyCase.loc[mask_HC]
HC_cases_mask
```

```python
HC_cases_mask.rename(columns = {'Last update date':'Date'}, inplace = True)
HC_cases_mask
```

```python
HC_cases_mask = HC_cases_mask.drop(['Confirmed cases'], axis=1)
```

```python
HC_cases_mask
```

```python
HC_cases_mask.to_csv("./HC_cases_mask.csv") 
```

```python
#set Date as index
HC_cases_mask_indexDate = HC_cases_mask.set_index('Date')
```

```python
HC_daily_case = plt.figure(figsize =(15,6))
HC_daily_case = HC_cases_mask_indexDate['Daily Confirmed Cases'].plot()
plt.xlabel("Date YYYY-MM")
plt.ylabel("Confirmed_Daily_Cases")
plt.legend()
plt.title('Confirmed Daily Case of COVID-19-Hartford Town,Connecticut')
```

Read Crime Data

```python
crime_raw = pd.read_csv('Police_Incidents_01012005_to_Current.csv') 
crime_raw
```

```python
#convert date type
crime_raw['Date'] = pd.to_datetime(crime_raw['Date'], format = '%m/%d/%Y')
```

```python
crime_raw['Date'].dtypes
```

```python
#sort by Date
crime_sort_date = crime_raw.sort_values(by = 'Date')
```

```python
crime_sort_date
```

```python
#get the data during Mask mandate and togther with HC_cases
mask = (crime_sort_date['Date'] >= '2020-06-01') &(crime_sort_date['Date'] <= '2021-05-31') 
crime_2020=crime_sort_date.loc[mask]
crime_2020
```

```python
crime_2020.to_csv("./crime_2020.csv") 
```

```python
#drop unused column
crime_2020_clean = crime_2020.drop(['Time_24HR','Address','UCR_2_Category', 'UCR_2_Description', 'UCR_2_Code'], axis=1)
```

```python
#Group by Date 
dailyCrime_2020 = crime_2020_clean.groupby('Date').count()
dailyCrime_2020 = dailyCrime_2020.drop(['UCR_1_Category', 'UCR_1_Description', 'UCR_1_Code', 'Neighborhood','geom'], axis=1)
```

```python
dailyCrime_2020.to_csv("./dailyCrime_2020.csv") 
```

```python
dailyCrime_2020
```

```python
dailyCrime_2020=dailyCrime_2020.reset_index()
dailyCrime_2020
```

```python
dailyCrime_2020.to_csv("./dailyCrime_2020.csv") 
```

```python
# Merge cases table and crime table together
HC_case_crime = dailyCrime_2020.merge(HC_cases_mask,left_on='Date', right_on='Date', how='outer')
HC_case_crime
```

```python
#convert Nan to 0
HC_case_crime = HC_case_crime.fillna(0)
HC_case_crime
```

```python
HC_case_crime.to_csv("./HC_case_crime_2020.csv") 
```

```python
Nov_Feb = (HC_case_crime['Date'] >= '2020-11-01') &(HC_case_crime['Date'] < '2021-03-01') 
HC_special=HC_case_crime.loc[Nov_Feb]
HC_special
```

```python
HC_special.to_csv("./HC_special.csv") 
```
